```python
import pandas as pd
```

```python
df = pd.read_csv("SuperStoreOrders.csv")
print(df)
```

```python
df.info()
```

```python
print(df.isnull().sum()) # check for missing values 
# so there is no missing values 
```

```python
# df['profit'].fillna(0, inplace=True)     it raises a error   # Fill missing profit values with 0]

df['profit'] = df['profit'].fillna(0)

print(df.info)
```

```python
df.head(15)
```

```python
# df["order_date"] = pd.to_datetime(df["order_date"])  default  --- mm/dd/yy or yy/mm/dd
# df["order_date"] = pd.to_datetime(df["order_date"], dayfirst=True)   --- it tell pandas that first part is day not month 
df["order_date"] = pd.to_datetime(df["order_date"], format="mixed", dayfirst=True)
df["ship_date"] = pd.to_datetime(df["ship_date"], format="mixed", dayfirst=True)
# first we have to remove , and $ etc.
df["sales"] = df["sales"].str.replace(",","").str.replace("$","")
df['sales'] = pd.to_numeric(df['sales'])
print(df)  
```

```python
print(df.dtypes)
```

```python
df.describe()    
```

```python
# to find the Selling¬†Price¬†per¬†Unit using formula sales/quantity
df["Selling¬†Price¬†per¬†Unit"] = df["sales"] / df["quantity"]
print(df.head,df.dtypes)

# print(df.dtypes)
```

```python
df.describe()
# there is no need to handle outliers like -- dicsount is b/w 0-1,  and -- sales are appropriate and -- profit also 
```

```python
# Standardize column names (just for using these commands)
# [ bcz all columns were alread  in lower case and with no space , there was already a '_' b/w them]
# df.columns = df.columns.str.lower().str.replace(" ", "_")
# df.rename(columns={"selling price per unit":"selling_price_per_unit"}, inplace=True
df.columns = df.columns.str.strip().str.lower().str.replace(" ", "_")
df.columns
# it is not werking for selling price per unit --------
```

```python
print(df.columns.tolist())  #Print column names with their exact representation
```

```python
df.columns = df.columns.str.strip().str.replace(r"\s+", "_", regex=True).str.lower()
"""  .str.strip() ‚Üí Removes leading and trailing spaces.
 .str.replace(r"\s+", "_", regex=True) ‚Üí Replaces all types of spaces (normal spaces, tabs, non-breaking spaces, etc.) with underscores  """
df.columns
```

```python
# Save cleaned data to a new CSV file
df.to_csv("cleaned_Super_store_orders_data.csv", index=False)

print("‚úÖ Data cleaning completed! Cleaned file saved as 'cleaned_Super_store_orders_data.csv'.")
```
